```python
import numpy as np
import csv

dataPath = '/mnt/c/users/llave/Documents/flow'
fname_sim='/sim/meanSpeeds_sim.csv'
fname_trained ='/simplified_envV3/meanSpeeds_trained.csv'
# fname_trained2 ='/simplified/meanSpeeds_simplified_c260.csv'
fname_trained2 = '/simplified_envV2/meanSpeeds_trained.csv'
# fname_trained3 = '/simplified_envV3/meanSpeedsV3.csv'

def import_data(fname):
    
    fields = []
    rows = []
    
    # reading csv file 
    with open(fname, 'r') as csvfile: 
        # creating a csv reader object 
        csvreader = csv.reader(csvfile) 

        # extracting field names through first row 
        fields = next(csvreader) 

        # extracting each data row one by one 
        for row in csvreader: 
            rows.append(float(row[0]))
    
    return fields, rows
```

```python
sim_fields, sim_rows = import_data(dataPath+fname_sim)
trained_fields, trained_rows = import_data(dataPath+fname_trained)
trained_fields2, trained_rows2 = import_data(dataPath+fname_trained2)
#trained_fields3, trained_rows3 = import_data(dataPath+fname_trained3)
```
